Question 3

Create a numpy array holding values 10 to 100 and going up in increments of 10. Slice 4th element through 9th element, incrementing by 2 as sub1. Then slice 1st element through 6th element, once again ncrementing by 2 as sub2. Multiply sub1 and sub2 and save result in 'product' and print the product. Finally print the square root of sub2.

```python
array = np.arange(10, 110, 10)

sub1 = array[3:9:2]

sub2 = array[0:6:2]

product = sub1 * sub2

print(product)

print(np.sqrt(sub2))
```

Question 4

Create a list called 'students' with the words 'Algebra I', 'Intro to Business', and 'Creative Writing'. Then create the series 
print the initial series with the values 30, 31, and 29. Create a panda series with classes as the index, then add a new element 'Programming' with the value of 31. Finally print only the elements where att is less than 31.

Question 6

Create a dictionary called salesDict holding cell phone data and a string holding dates: '01/01/2020', '01/02/2020', '01/03/2020', '01/04/2020'. Use salesDict and dates to create a dataframe called sales and print sales. Create a Mean, Median, and Mode column using the data from sales and drop the columns with a phone name.

```python
salesDict = {'Samsung Galaxy S10': [769.34, 834.23, 900.12, 1021.12],
'iPhone X': [983.11, 881.21, 1210.32, 1100.34],
'Google Pixel 4': [1021.18, 1321.12, 832.14, 992.15]}
dates=(['01/01/2020','01/02/2020','01/03/2020','01/04/2020'])
phoneDF= pd.DataFrame(salesDict, index=[dates])
print(phoneDF)

sales=pd.DataFrame(salesDict, index=[dates])
sales['Mean']=sales.mean(axis=1)
sales['Median']=sales.median(axis=1)
sales['St Dev']=sales.std(axis=1)
sales=sales[['Mean','Median','St Dev']]
print(sales)
```

Question 7

The code reads a CSV file "SalesJan2009.csv" into a Pandas DataFrame salesjan. It then creates a bar plot to display the mean price by payment type. The x-axis shows the payment types and the y-axis shows the mean price. Finally, the plot is displayed using plt.show().

```python
import seaborn as sns
import matplotlib.pyplot as plt

salesjan = pd.read_csv("Data/SalesJan2009.csv", sep=',')

sns.set(style='whitegrid')
plt.figure(figsize=(15, 5))
ax = sns.barplot(salesjan.Payment_Type, salesjan.Price)
ax.set(title='Mean Price by Payment Type', xlabel='Payment Type', ylabel='Mean Price')
plt.show()
```

Question 8

Load the SalesJan2009.csv dataset into a Pandas DataFrame and create histogram of 'Price' with 5 bins and an axis limit of 0 to 800. Set the plot title, axis lables, and show the plot.

```python
import matplotlib.pyplot as plt

salesjan = pd.read_csv("Data/SalesJan2009.csv", sep=',')

plt.hist(salesjan['Price'], bins=5)

plt.xlim(0, 8000)

plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Count')

plt.show()
```
